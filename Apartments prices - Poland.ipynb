## Apartments prices in Poland

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
apartments = pd.read_csv("apartments_pl_2023_08.csv")
apartments.head()
```

```python
apartments.info()
```

### Data Exploration

### Average prices in different cities 

```python
average_prices_by_city = apartments.groupby('city')['price'].mean()
print("Average prices of apartments in each city:")
for city, price in average_prices_by_city.items():
    print(f"{city}: {price:.2f}zÅ‚")
```

```python
plt.bar(average_prices_by_city.index, average_prices_by_city)
plt.xlabel('City')
plt.ylabel('Average Price')
plt.title('Average Apartment Prices by City')
plt.xticks(rotation=50)
plt.show()
```

### Average price for one square meter

```python
apartments['price_for_square_meter'] = apartments['price']/apartments['squareMeters']
apartments.head()['price_for_square_meter']
```

```python
avg_price_for_square_meter = apartments.groupby('city')['price_for_square_meter'].mean()
avg_price_for_square_meter
```

```python
plt.bar(avg_price_for_square_meter.index, avg_price_for_square_meter)
plt.xlabel('City')
plt.ylabel('Average Price for square meter')
plt.title('Average prices for square meter')
plt.xticks(rotation=50)
plt.show()
```

### Quartiles of prices in different cities

```python
quartiles_by_city = apartments.groupby('city')['price'].quantile([0.25, 0.50, 0.75])
print("Quartiles by city:")
quartiles_by_city
```

```python
cities=list(apartments['city'].unique())
cities.sort()
for c in cities:

    plt.figure()
    plt.boxplot(apartments[apartments['city']==c]['price'])
    plt.title(c)
    plt.ylabel('Price')
    
    plt.show()
```

### Quartiles by square meter

```python
quartiles_by_square_meter = apartments.groupby('city')['price_for_square_meter'].quantile([0.25, 0.50, 0.75])
print("Quartiles by square meter:")
quartiles_by_square_meter
```

```python
for c in cities:

    plt.figure()
    plt.boxplot(apartments[apartments['city']==c]['price_for_square_meter'])
    plt.title(c)
    plt.ylabel('Price for square meter')
    
    plt.show()
```

### Square meter price vs. size of apartment

```python
plt.scatter(apartments['squareMeters'], apartments['price_for_square_meter'],marker='.')
plt.xlabel('Size')
plt.ylabel('Price')
plt.title('Square meter price vs. size of apartment')
plt.grid()
plt.show()
```

##### Question 1: How big apartment can we buy for 500 000 zÅ‚ in different cities?

##### Average size for 500 000zÅ‚

```python
affordable_apartments = apartments[apartments['price'] <= 500000]
avg_affordable_apartments_by_city = affordable_apartments.groupby('city')['squareMeters'].mean()
avg_affordable_apartments_by_city
```

```python
plt.bar(avg_affordable_apartments_by_city.index, avg_affordable_apartments_by_city)
plt.xlabel('City')
plt.ylabel('Square Meters')
plt.title('Average size of apartment under 500 000zÅ‚ in different cities')
plt.xticks(rotation=50)
plt.show()
```

##### The biggest size for 500 000zÅ‚


```python
max_affordable_apartments_by_city = affordable_apartments.groupby('city')['squareMeters'].max()
max_affordable_apartments_by_city
```

```python
plt.bar(max_affordable_apartments_by_city.index, max_affordable_apartments_by_city)
plt.xlabel('City')
plt.ylabel('Square Meters')
plt.title('The biggest size of apartment under 500 000zÅ‚ in different cities')
plt.xticks(rotation=50)
plt.show()
```

##### Question 3: How expensive are flats from different years?

##### Question 4:  How expensive are one, two and three-rooms apartment in different cities?

```python
one_room_apartments = apartments[apartments['rooms'] == 1]
two_rooms_apartments = apartments[apartments['rooms'] == 2]
three_rooms_apartments = apartments[apartments['rooms'] == 3]

avg_one_room_price = one_room_apartments.groupby('city')['price'].mean()
avg_two_rooms_price = two_rooms_apartments.groupby('city')['price'].mean()
avg_three_rooms_price = three_rooms_apartments.groupby('city')['price'].mean()

print("Average price for one room: \n", avg_one_room_price, "\n")
avg_one_room_price

print("Average price for two rooms: \n",avg_two_rooms_price, "\n")


print("Average price for three rooms: \n", avg_three_rooms_price, "\n")
```

```python
cities = avg_one_room_price.index
bar_width = 0.25

x = np.arange(len(cities))

fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(x - bar_width, avg_one_room_price, width=bar_width, label='One-Room')
ax.bar(x, avg_two_rooms_price, width=bar_width, label='Two-Room')
ax.bar(x + bar_width, avg_three_rooms_price, width=bar_width, label='Three-Room')

ax.set_xticks(x)
ax.set_xticklabels(cities, rotation=45)
ax.set_xlabel('City')

ax.set_ylabel('Average Price')
ax.set_title('Average Apartment Prices by City and Room Type')

ax.legend()

plt.tight_layout()
plt.show()
```
